```python
data=data.interpolate()
# data.isnull().sum()
```

```python
data.engine =data.engine.str.replace("CC"," ")
```

```python
data.info()
```

```python
x=data.engine.fillna("ck")
x=list(x)
```

```python
for i in range(len(x)):
    if x[i]=="ck":
        x[i]=x[i-1]+" "
```

```python
for i in range(len(x)):
    x[i]=x[i]+"CC"
```

```python
data.engine = x
```

```python
# data
# data.info()
data.isnull().sum()
```

```python
data.mileage=data.mileage.str.replace("kmpl"," ")
```

```python
r=data.mileage.fillna("ck")

# t1=data["mileage"].to_numpy()# intake values
t1=list(r)
for i in range(len(t1)):
    if t1[i]=="ck":
        t1[i]=t1[i-1]
```

```python
for i in range(len(t1)):
    t1[i]=t1[i]+"kmph"
    
data.mileage=t1
```
